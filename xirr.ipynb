```python
def xnpv(rate, cashflows):
    return sum([cf/(1+rate)**((t-cashflows[0][0]).days/365.0) for (t,cf) in cashflows])
 
def xirr(cashflows, guess=0.1):
    try:
        return optimize.newton(lambda r: xnpv(r,cashflows),guess)
    except:
        print('Calc Wrong')

#data = [(datetime.date(2021, 6, 1), -4000), (datetime.date(2021, 7, 1), 13.33333), (datetime.date(2021, 8, 1), 13.33333), (datetime.date(2021, 9, 1), 13.33333), (datetime.date(2021, 10, 1), 13.33333), (datetime.date(2021, 11, 1), 13.33333), (datetime.date(2021, 12, 1), 13.33333), (datetime.date(2022, 1, 1), 13.33333), (datetime.date(2022, 2, 1), 13.33333), (datetime.date(2022, 3, 1), 13.33333), (datetime.date(2022, 4, 1), 13.33333), (datetime.date(2022, 5, 1), 13.33333), (datetime.date(2022, 6, 1), 4013.33333)]
data = list(zip(df['Date'].values,df['Flow'].values))
# keep in mind that this is the annual ineternal rate of return and you can have that multiplied by number of year.
xirr(data)
```
